```python
# Download data from http://goo.gl/j1MSKe (=> imf.org)
gdp_per_capita = pd.read_csv(datapath+"gdp_per_capita.csv", thousands=',', delimiter='\t',
                             encoding='latin1', na_values="n/a")
gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp_per_capita.set_index("Country", inplace=True)
gdp_per_capita.head(2)
```

```python
full_stat=pd.merge(left=oecd_bli,right=gdp_per_capita,left_index=True,right_index=True)
full_stat.sort_values(by="GDP per capita", inplace=True)
full_stat.shape
oecd_bli.shape
```

```python
from sklearn import linear_model
lin1 = linear_model.LinearRegression()
lin2 = linear_model.LinearRegression()
gdp=full_stat["GDP per capita"]
print(type(gdp),gdp.shape)

Xsample = full_stat["GDP per capita"].values.reshape(-1,1)
Xsample2 = full_stat["Personal earnings"].values.reshape(-1,1)
ysample = full_stat["Life satisfaction"].values.reshape(-1,1)
# Xsample = full_stat["GDP per capita"]
# ysample = full_stat["Life satisfaction"]

lin1.fit(Xsample, ysample)
lin2.fit(Xsample2, ysample)
```

```python
from sklearn.metrics import mean_squared_error
pred1_dgp=lin1.predict(Xsample)
pred2_ern=lin1.predict(Xsample2)
err1=mean_squared_error(ysample,pred1_dgp)
err2=mean_squared_error(ysample,pred2_ern)

print(np.sqrt(err1),np.sqrt(err2))
```
